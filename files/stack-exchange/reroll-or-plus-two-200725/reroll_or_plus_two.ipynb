```python
from dyce import H
from fractions import Fraction
import pandas

d_fudge = 4 @ H((-1, 0, 1))
DATA_COLUMNS = [f"Roll {outcome}" for outcome in d_fudge]
LADDER = list(range(-2, 7))

df = pandas.DataFrame(columns=DATA_COLUMNS)

for target_val in LADDER:
    row_data = []
    for initial_outcome, initial_count in d_fudge.items():
        plus_two_res = Fraction(1) if initial_outcome + 2 >= target_val else Fraction(0)
        reroll_h = d_fudge.ge(target_val)
        reroll_res = Fraction(reroll_h.get(True, 0), reroll_h.total)
        if initial_outcome >= target_val:
            advice = "Stop"
            res = plus_two_res
        elif plus_two_res > reroll_res:
            advice = "+2"
            res = plus_two_res
        elif plus_two_res < reroll_res:
            advice = "Reroll"
            res = reroll_res
        else:
            advice = "Either"
            res = plus_two_res
        res_fmt = f"{float(res):.0%}" if res == 0 or res == 1 else f"{float(res):.2%}"
        row_data.append(f"{advice} ({res_fmt})")
    row = pandas.DataFrame([row_data], columns=DATA_COLUMNS, index=[target_val])
    df = pandas.concat((df, row))

df.index.name = "Difficulty"
display(df)
# print(df.to_markdown())
```
